```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

# The following block will open FireFox and crawl the first 10 "pages" of FDs latest news

```python
url = "https://fd.nl/laatste-nieuws"

driver = webdriver.Firefox()
driver.implicitly_wait(20)
driver.get(url)

python_button = driver.find_elements_by_xpath("//*[@id='cookieconsent1']")[0]

python_button.click()

for _ in range(20):
    try:
        pb = driver.find_elements_by_xpath("/html/body/div[3]/div/section/main/div/button")[0]
        pb.click()
    except:
        break
html = driver.page_source.encode('utf-8')

```

```python
soup = BeautifulSoup(html, 'lxml')

topics_dict = { "title":[], \
                "sub":[], \
                "link":[], \
                "created":[], \
                }

topics_data_old = pd.read_csv("FD-News.csv")

content_blocks = soup.find_all("article")
x = 0
for cb in content_blocks:
    x = x + 1
    #print(x)
    try:
        title = cb.h1.text
        sub = cb.p.text
        href = cb.a.get('href')
        link = "https://fd.nl" + str(href)
        created = cb.time.text

        title.replace(",",";")
        sub.replace(",",";")


        topics_dict["title"].append(title)
        topics_dict["sub"].append(sub)
        topics_dict["link"].append(link)
        topics_dict["created"].append(created)
    except:
        print('failed')
        pass

# content_blocks_art = soup.find_all("article")
# for cb in content_blocks_art:
#     x = x + 1
#     #print(x)
#     try:
#         title = cb.h1.text
#         sub = cb.p.text
#         href = cb.a.get('href')
#         link = "https://fd.nl" + str(href)
#         created = cb.time.text
        
#         title.replace(",",";")
#         sub.replace(",",";")
        
        
#         topics_dict["title"].append(title)
#         topics_dict["sub"].append(sub)
#         topics_dict["link"].append(link)
#         topics_dict["created"].append(created)
#     except:
#         print('failed')
#         pass

print("DONE")
```

```python
topics_data_new = pd.DataFrame(topics_dict)
topics_data_all = topics_data_old.append(topics_data_new, sort=False, ignore_index=True)

test = topics_data_old.drop_duplicates(['link'])

topics_data_all = topics_data_all.drop_duplicates(['link'],keep='last')

print(len(topics_data_new))
print(len(topics_data_all))
print(len(topics_data_old))
print(len(test))

topics_data_all.to_csv("FD-News.csv", encoding = "utf-8", index= False)
```
